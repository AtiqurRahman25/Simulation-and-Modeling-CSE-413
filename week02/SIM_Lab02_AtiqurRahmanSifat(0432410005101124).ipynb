```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

### Assingment 01 - Basic Matrix Operations

```python
# Create two 3Ã—3 matrices with random integers in Python.
m1 = np.random.randint(0, 10, size = (3, 3))
m2 = np.random.randint(0, 10, size = (3, 3))
print(m1)
print(m2)
```

```python
# Perform addition, subtraction, and multiplication operations on these matrices.

addM = m1 + m2
subM = m1 - m2
dotM = np.dot(m1, m2)

print(addM)
print(subM)
print(dotM)
```

```python
# Compute the determinant, inverse, and rank of one of the matrices.

# Determinant of a matrix
detM1 = np.linalg.det(m1)

# Inverse of a matrix
if np.linalg.det(m1) != 0:  # Check if matrix is invertible
    invM1 = np.linalg.inv(m1)  # Compute the inverse of A
else:
    invM1 = "Matrix is singular, cannot compute inverse"  # No inverse if determinant is 0

#  Rank of a matrix
rankM1 = np.linalg.matrix_rank(m1)

print(detM1)
print(invM1)
print(rankM1)
```

### Assignment 2: Matrix Plotting and Visualization

1. Generate and Plot Two Random Vectors:


```python
# Create two vectors, each containing 15 random floats.
vector1 = np.random.rand(15)
vector2 = np.random.rand(15)

# Plot the vectors
plt.plot(vector1, label='Vector 1', marker='o')
plt.plot(vector2, label='Vector 2', marker='s')

# Label the axes
plt.xlabel('Index')
plt.ylabel('Value')

# Title of the plot
plt.title('Comparison of Two Random Float Vectors')

# Show legend
plt.legend()

# Show plot
plt.show()
```

2. Generate and Visualize a Random Matrix:

```python

# 4x4 matrix with random values
matrix = np.random.rand(4,4)

# row and column labels
row_labels = [f'Row {i}' for i in range(1, 5)]
col_labels = [f'Col {i}' for i in range(1, 5)]

# Plot heatmap
plt.figure(figsize=(6,5))
sns.heatmap(matrix, annot=True, cmap='viridis', xticklabels=col_labels, yticklabels=row_labels)
plt.title("4x4 Random Matrix Heatmap")
plt.show()
```

3. Matrix Operations and Visualization:

```python
# 4x4 matrix with random values
matrix1 = np.random.rand(4,4)
matrix2 = np.random.rand(4,4)


# Perform the following operations:
# Matrix Addition
# Matrix Subtraction
# Matrix Multiplication (ensure matrices are compatible for multiplication)

addM = matrix1 + matrix2
subM = matrix1 - matrix2
dotM = np.dot(matrix1, matrix2)

# print(addM)
# print(subM)
# print(dotM)

# For each of these operations, visualize the result using bar plots. Each bar should represent one element of the matrix.

# Bar plot to visualize results
fig, axes = plt.subplots(1, 3, figsize=(15, 5)) #figure object, subplot array

# Sum
axes[0].bar(range(1, 17), addM.flatten(), color='b')
axes[0].set_title('Matrix Addition')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Value')

# Difference
axes[1].bar(range(1, 17), subM.flatten(), color='g')
axes[1].set_title('Matrix Subtraction')
axes[1].set_xlabel('Index')
axes[1].set_ylabel('Value')

# Product
axes[2].bar(range(1, 17), dotM.flatten(), color='r')
axes[2].set_title('Matrix Multiplication')
axes[2].set_xlabel('Index')
axes[2].set_ylabel('Value')

plt.tight_layout() #plt.tight_layout() is a Matplotlib function that automatically adjusts the layout of subplots and figures to prevent overlapping and ensure a visually appealing presentation. It modifies the spacing between subplots, axes, labels, and titles to improve readability.
# plt.show()
```
